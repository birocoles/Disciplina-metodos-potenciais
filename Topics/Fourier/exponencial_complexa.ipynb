```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 721)
```

```python
N = x.size
n = np.arange(N)

f = (1.*n)/(N-1)

e = np.exp(-2j*f)
```

```python
y = 3.*np.cos(x) + 5.*np.sin(x)
```

```python
Y = y*e
```

```python
ymax = np.max(np.abs([np.min(y), np.max(y)]))
```

```python
plt.figure(figsize=(14,14))

plt.subplot(3,3,1)
plt.plot(x,y,'k.')

plt.subplot(3,3,2)
plt.plot(f,e.real,'k.')

plt.subplot(3,3,3)
plt.plot(f,e.imag,'k.')

plt.subplot(3,3,4)
plt.plot(f,Y.real,'k.')
plt.plot(f,e.real,'k--')
plt.plot(f,y,'r-')

plt.subplot(3,3,5)
plt.plot(f,Y.imag,'k.')
plt.plot(f,e.imag,'k--')
plt.plot(f,y,'r-')

plt.subplot(3,3,6)
plt.plot(f,np.abs(Y.real),'k.')
plt.plot(f,np.abs(e.real)*ymax,'k--')
plt.plot(f,np.abs(y),'r-')

plt.subplot(3,3,7)
plt.plot(f,np.abs(Y.imag),'k.')
plt.plot(f,np.abs(e.imag)*ymax,'k--')
plt.plot(f,np.abs(y),'r-')

plt.show()
```
