```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H0 = np.ones(3)
```

```python
chi_values = np.linspace(0.001, 1, 100)
```

```python
I = np.identity(3)
demag = 1/3
```

```python
h_true_norm = []
h_approx_norm = []
for chi in chi_values:
    # compute h_approx_norm
    h_approx = chi*H0
    h_approx_norm.append(np.linalg.norm(h_approx, ord=2))
    # compute h_true_norm
    aux = np.linalg.inv(I + demag*chi*I)
    h_true = np.dot(aux, h_approx)
    h_true_norm.append(np.linalg.norm(h_true, ord=2))
h_true_norm = np.array(h_true_norm)
h_approx_norm = np.array(h_approx_norm)
```

```python
diff = h_true_norm - h_approx_norm
```

```python
plt.figure()
plt.plot(chi_values, h_true_norm, 'b-', label='true')
plt.plot(chi_values, h_approx_norm, 'r-', label='approx')
plt.plot(chi_values, diff, 'k--', label='true - approx')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```
