```python
%matplotlib inline
import numpy as np
from IPython.display import Image

from fatiando import mesher, gridder
from fatiando.gravmag import prism
from fatiando.utils import mgal2si, ang2vec
from fatiando.vis import mpl, myv
from fatiando.constants import G, CM
```

## Cubo sintético

```python
Lp = 2000.0 # aresta do cubo em metros

# cubo
model = [mesher.Prism(-0.5*Lp, 0.5*Lp, -0.5*Lp, 0.5*Lp, 0, Lp)]

bounds = [-Lp, Lp, -Lp, Lp, 0, 2*Lp]

# Visualização do corpo
scene = myv.figure(size=(600,600))
myv.prisms(model, color=(0,0,1))
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=5, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.5
myv.wall_bottom(ax.axes.bounds, opacity=0.2)
myv.wall_north(ax.axes.bounds)
scene.scene.camera.position = [-9104.4101911922098, -10043.828578857845, -2342.4591286514151]
scene.scene.camera.focal_point = [8.9157087444868326e-12, 8.9157087444868326e-12, 1999.9999999999975]
scene.scene.camera.view_angle = 30.0
scene.scene.camera.view_up = [0.20215054673967989, 0.22849603665955914, -0.95232595138624765]
scene.scene.camera.clipping_range = [7524.3416656038644, 22715.917681673785]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('cubo.png')
myv.show()
Image(filename='cubo.png')
```

## Propriedades físicas do corpo sintético

```python
inc, dec = -30.0, -13.0 # inclinação e declinação do campo geomagnético
incs, decs = 18.0, 24.0 # inclinação e declinação do vetor magnetização do corpo
ints = 6.0 # intensidade do vetor magnetização do corpo

m_versor = ang2vec(1.0,incs,decs) # versor do vetor de magnetização
```

```python
rho = 1000 # densidade do corpo em kg/m³
```

```python
model[0].addprop('magnetization', ints*m_versor)
model[0].addprop('density', rho)
```

## Cálculo das componentes Cartesianas da indução magnética e do tensor gradiente da gravidade produzida pelo corpo sintético em um plano horizontal

```python
xmin = -0.5*10*Lp
xmax =  0.5*10*Lp
ymin = -0.5*10*Lp
ymax =  0.5*10*Lp

area = [xmin, xmax, ymin, ymax]

nx = 2001
ny = 2001
n = nx*ny
shape = (ny, nx)

x, y, z = gridder.regular(area, shape, z=-0.1*Lp) # coordenadas dos dados

gxx = prism.gxx(x, y, z, model) # componente xx do tensor gradiente gravitacional
gxy = prism.gxy(x, y, z, model) # componente xy do tensor gradiente gravitacional
gxz = prism.gxz(x, y, z, model) # componente xz do tensor gradiente gravitacional
gyy = prism.gyy(x, y, z, model) # componente yy do tensor gradiente gravitacional
gyz = prism.gyz(x, y, z, model) # componente yz do tensor gradiente gravitacional
gzz = prism.gzz(x, y, z, model) # componente zz do tensor gradiente gravitacional
bx = prism.bx(x, y, z, model)   # componente x do vetor indução magnética
by = prism.by(x, y, z, model)   # componente y do vetor indução magnética
bz = prism.bz(x, y, z, model)   # componente z do vetor indução magnética
```

## Relação de Poisson

Seja um corpo com volume $v$, densidade $\rho$ e vetor magnetização $\mathbf{m}$. De acordo com a **Relação de Poisson**, 
a indução magnética $\mathbf{B}(x,y,z)$ e o tensor gradiente da gravidade $\mathbf{T}(x,y,z)$ preditos por este corpo são satisfazem a seguinte equação:

$\mathbf{B}(x,y,z)=\dfrac{C_{m} \, m}{G \, \rho} \, \mathbf{T}(x,y,z) \, \hat{\mathbf{m}}$ ,

em que $C_{m} = \frac{\mu_{0}}{4 \, \pi} = 10^{-7} \frac{H}{m}$, $\mu_{0}$ é a permeabilidade magnética no vácuo, $G$ é a constante gravitacional, $\hat{\mathbf{m}}$ é um vetor unitário com a mesma direção e sentido de $\mathbf{m}$, $m$ é a intensidade de $\mathbf{m}$ e

$\mathbf{T}(x,y,z) = G \, \rho \, \left[
\begin{array}{ccc}
\phi_{xx}(x,y,z) & \phi_{xy}(x,y,z) & \phi_{xz}(x,y,z) \\
\phi_{xy}(x,y,z) & \phi_{yy}(x,y,z) & \phi_{yz}(x,y,z) \\
\phi_{xz}(x,y,z) & \phi_{yz}(x,y,z) & \phi_{zz}(x,y,z) \\
\end{array} \right]$ ,

sendo $\phi_{\alpha \beta}$, $\alpha = x, y, z$, $\beta = x, y, z$, as segundas derivadas da função

$\phi(x,y,z) = \iiint_{v} \dfrac{1}{\sqrt{(x-x')^{2}+(y-y')^{2}+(z-z')^{2}}} dx'\,dy'\,dz'$

em relação às coordenadas $x$, $y$ e $z$ do ponto de observação.

```python
constante_poisson = ((CM*ints)/(G*rho))
bx_poisson = constante_poisson*((gxx*m_versor[0])+(gxy*m_versor[1])+(gxz*m_versor[2]))
by_poisson = constante_poisson*((gxy*m_versor[0])+(gyy*m_versor[1])+(gyz*m_versor[2]))
bz_poisson = constante_poisson*((gxz*m_versor[0])+(gyz*m_versor[1])+(gzz*m_versor[2]))
```

### Comparação entre as componentes $B_{x}$ calculadas diretamente e aquelas preditas pela *Relação de Poisson*

```python
mpl.close('all')
mpl.figure(figsize=(15, 5))
mpl.axis('scaled')

mpl.subplot(1,2,1)
mpl.title('$B_{x}$ (nT)', fontsize=20)
levels = mpl.contourf(y, x, bx, shape, 15)
cb = mpl.colorbar(shrink=0.8)
mpl.contour(y, x, bx, shape,
            levels, clabel=False, linewidth=0.1)
mpl.xlabel('$y$ (km)', fontsize=18)
mpl.ylabel('$x$ (km)', fontsize=18)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.subplot(1,2,2)
mpl.title('$B_{x}$ (nT) - Poisson', fontsize=20)
levels = mpl.contourf(y, x, bx_poisson, shape, 15)
cb = mpl.colorbar(shrink=0.8)
mpl.contour(y, x, bx_poisson, shape,
            levels, clabel=False, linewidth=0.1)
mpl.xlabel('$y$ (km)', fontsize=18)
mpl.ylabel('$x$ (km)', fontsize=18)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.show()
```

### Comparação entre as componentes $B_{y}$ calculadas diretamente e aquelas preditas pela *Relação de Poisson*

```python
mpl.close('all')
mpl.figure(figsize=(15, 5))
mpl.axis('scaled')

mpl.subplot(1,2,1)
mpl.title('$B_{y}$ (nT)', fontsize=20)
levels = mpl.contourf(y, x, by, shape, 15)
cb = mpl.colorbar(shrink=0.8)
mpl.contour(y, x, by, shape,
            levels, clabel=False, linewidth=0.1)
mpl.xlabel('$y$ (km)', fontsize=18)
mpl.ylabel('$x$ (km)', fontsize=18)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.subplot(1,2,2)
mpl.title('$B_{y}$ (nT) - Poisson', fontsize=20)
levels = mpl.contourf(y, x, by_poisson, shape, 15)
cb = mpl.colorbar(shrink=0.8)
mpl.contour(y, x, by_poisson, shape,
            levels, clabel=False, linewidth=0.1)
mpl.xlabel('$y$ (km)', fontsize=18)
mpl.ylabel('$x$ (km)', fontsize=18)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.show()
```

```python
mpl.close('all')
mpl.figure(figsize=(15, 5))
mpl.axis('scaled')

mpl.subplot(1,2,1)
mpl.title('$B_{z}$ (nT)', fontsize=20)
levels = mpl.contourf(y, x, bz, shape, 15)
cb = mpl.colorbar(shrink=0.8)
mpl.contour(y, x, bz, shape,
            levels, clabel=False, linewidth=0.1)
mpl.xlabel('$y$ (km)', fontsize=18)
mpl.ylabel('$x$ (km)', fontsize=18)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.subplot(1,2,2)
mpl.title('$B_{z}$ (nT) - Poisson', fontsize=20)
levels = mpl.contourf(y, x, bz_poisson, shape, 15)
cb = mpl.colorbar(shrink=0.8)
mpl.contour(y, x, bz_poisson, shape,
            levels, clabel=False, linewidth=0.1)
mpl.xlabel('$y$ (km)', fontsize=18)
mpl.ylabel('$x$ (km)', fontsize=18)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.show()
```
