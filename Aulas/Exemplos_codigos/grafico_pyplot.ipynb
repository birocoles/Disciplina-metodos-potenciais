# Como plotar o gráfico de uma função $f(x)$?

Este código exemplifica como gerar o gráfico de uma função de uma variável utilizando o `pyplot`. 

## Coisas para importar

```python
import numpy as np
from matplotlib import pyplot as plt
```

O comando executado na célula abaixo faz com que os gráficos sejam inseridos no próprio `notebook`.

```python
%matplotlib inline
```

## Parâmetros de entrada

```python
x = np.linspace(-30., 30., 61)
```

Observe que `x` é um [`numpy array`](http://docs.scipy.org/doc/numpy/reference/generated/numpy.array.html) 1D criado a partir da função [`linspace`](http://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html) do `numpy.`

```python
f = x**2.
```

O comando acima calcula um `numpy array` `f` a partir dos elementos do `numpy array` `x`. Cada elemento de `f` é igual ao quadrado do respectivo elemento de `x`.

## Exemplos de gráficos feitos com o [`pyplot`](http://matplotlib.org/api/pyplot_api.html)

```python
plt.figure(figsize=(8,6))                             #Comando para criar a figura
plt.plot(x, f, 'ko')                                  #Comando para plotar o gráfico
plt.title('funcao $f(x) = x^{2}$', fontsize = 20)     #Título
plt.xlabel('$x$', fontsize = 18)                      #Texto do eixo x (abcissa)
plt.ylabel('$f(x)$', fontsize = 18)                   #Texto do eixo f(x) (ordenada)
plt.xticks(fontsize = 14)                             #Tamanho dos números no eixo x
plt.yticks(fontsize = 14)                             #Tamanho dos números no eixo f(x)
plt.show()                                            #Comando para plotar a figura
```

```python
plt.figure(figsize=(8,6))                             #Comando para criar a figura
plt.plot(x, f, 'k-')                                  #Comando para plotar o gráfico
plt.title('funcao $f(x) = x^{2}$', fontsize = 20)     #Título
plt.xlabel('$x$', fontsize = 18)                      #Texto do eixo x (abcissa)
plt.ylabel('$f(x)$', fontsize = 18)                   #Texto do eixo f(x) (ordenada)
plt.xticks(fontsize = 14)                             #Tamanho dos números no eixo x
plt.yticks(fontsize = 14)                             #Tamanho dos números no eixo f(x)
plt.show()                                            #Comando para plotar a figura
```

```python
plt.figure(figsize=(8,6))                             #Comando para criar a figura
plt.plot(x, f, 'k-')                                  #Comando para plotar o gráfico
plt.plot(x, f, 'ko')                                  #Comando para plotar o gráfico
plt.title('funcao $f(x) = x^{2}$', fontsize = 20)     #Título
plt.xlabel('$x$', fontsize = 18)                      #Texto do eixo x (abcissa)
plt.ylabel('$f(x)$', fontsize = 18)                   #Texto do eixo f(x) (ordenada)
plt.xticks(fontsize = 14)                             #Tamanho dos números no eixo x
plt.yticks(fontsize = 14)                             #Tamanho dos números no eixo f(x)
plt.show()                                            #Comando para plotar a figura
```

Note que os gráficos foram feitos com diferentes estilos de linha e de marcadores. Para pesquisar diferentes tipos de marcadores e estilos de linha, veja o *help* da função `pyplot.plot`. Para tanto, execute o comando

```python
plt.plot?
```
