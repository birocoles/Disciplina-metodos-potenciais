```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from fatiando.constants import SI2MGAL
```

```python
def WGS84():
    '''
    This function returns the following parameters defining the 
    reference elipsoid WGS84:
    a = semimajor axis [m]
    f = flattening
    GM = geocentric gravitational constant of the Earth 
         (including the atmosphere) [m**3/s**-2]
    omega = angular velocity [rad/s]
    
    output:
    a, f, GM, omega
    '''
    a = 6378137.0
    f = 1.0/298.257223563
    GM = 3986004.418*(10**8)
    omega = 7292115*(10**-11)
    
    return a, f, GM, omega
```

```python
def elip2cart(h, lat, lon, a, f):

    b = a*(1.0-f)
    latr = np.deg2rad(lat)
    lonr = np.deg2rad(lon)
    N = (a**2)/np.sqrt((a*np.cos(latr))**2 + (b*np.sin(latr))**2)
    x = (N + h)*np.cos(latr)*np.cos(lonr)
    y = (N + h)*np.cos(latr)*np.sin(lonr)
    z = (N*(b/a)**2 + h)*np.sin(latr)
    
    return x, y, z
```

```python
def Rlat(x, y, z):
    return np.sqrt(x**2 + y**2 + z**2)
```

### $E = \left( \dfrac{R_{\phi} - h}{R_{\phi}^{2}} \right) \,  \left( 2 \, V_{\phi} \, V_{e} + V^{2} \right)$

### $E = 7.503 \, V \cos \phi \, \sin \alpha + 0.004154 \, V^{2}$

### Nas expressões acima, a velocidade $V$ está em $knot = 0.5144444 \, m/s$ e $E$ está em $mGal$

```python
def eotvos(Rlat, h, lat, omega, V, alpha):

    latr = np.deg2rad(lat)
    alphar = np.deg2rad(alpha)
    Vlat = Rlat*omega*np.cos(latr)
    Ve = V*np.sin(alphar)
    
    result = ((Rlat+h)/(Rlat**2))*(2*Vlat*Ve + V**2)*SI2MGAL
    
    #result = 7.503*V*np.cos(latr)*np.sin(alphar) + 0.004156*V**2
    
    return result
```

```python
def eotvosHarlan(h, lat, omega, V, alpha, a, f):
    
    latr = np.deg2rad(lat)
    alphar = np.deg2rad(alpha)
    coslatr = np.cos(latr)
    sinalphar = np.sin(alphar)
    
    #result = ((V**2)/a)*(1 + (h/a) - f*(1 - (coslatr**2)*(3 - 2*sinalphar**2)))
    #result += 2*V*omega*coslatr*sinalphar*(1+(h/a))*SI2MGAL

    result = ((V**2)/a)*(1 - (h/a) - f*(1 - (coslatr**2)*(3 - 2*sinalphar**2)))
    result += 2*V*omega*coslatr*sinalphar*SI2MGAL    
    
    return result
```

```python
lon = 0.0
h = 100.0
lat = np.linspace(0.0, 90.0, 91)
a, f, GM, omega = WGS84()
x, y, z = elip2cart(h, lat, lon, a, f)
```

```python
R = Rlat(x, y, z)
```

```python
print '%.5f %.5f %.5f %.5f' % (np.mean(R), np.std(R), R[0], R[-1])
```

```python
#Fórmula apresentada por Glicken (1962)
k3 = np.mean((R+h)/(R**2))
k2 = (0.514444444**2)*k3*SI2MGAL
k1 = 0.514444444*k3*2*np.mean(R)*omega*SI2MGAL

print '%.3f %.6f' % (k1,k2)
```

```python
alpha1 = 0.0 # degrees measured clockwise from true north
alpha2 = 10.0
alpha3 = 20.0
#V = 0.514444*300 # m/s
V = 150 # m/s
#V = 450 # knots
E1 = eotvos(R, h, lat, omega, V, alpha1)
E2 = eotvos(R, h, lat, omega, V, alpha2)
E3 = eotvos(R, h, lat, omega, V, alpha3)

Eh1 = eotvosHarlan(h, lat, omega, V, alpha1, a, f)
Eh2 = eotvosHarlan(h, lat, omega, V, alpha2, a, f)
Eh3 = eotvosHarlan(h, lat, omega, V, alpha3, a, f)
```

```python
plt.close('all')
plt.figure(figsize=(8,8))
plt.title('Approximated Eotvos correction (mGal)', fontsize=20)
plt.plot(lat,E1,'b-', label='$\\alpha$ = %.1f$^{\circ}$' % alpha1)
plt.plot(lat,E2,'r-', label='$\\alpha$ = %.1f$^{\circ}$' % alpha2)
plt.plot(lat,E3,'k-', label='$\\alpha$ = %.1f$^{\circ}$' % alpha3)
plt.plot(lat,Eh1,'b--', label='$\\alpha$ = %.1f$^{\circ}$' % alpha1)
plt.plot(lat,Eh2,'r--', label='$\\alpha$ = %.1f$^{\circ}$' % alpha2)
plt.plot(lat,Eh3,'k--', label='$\\alpha$ = %.1f$^{\circ}$' % alpha3)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('$\\phi \, (^{\circ})$', fontsize=20)
plt.ylabel('E (mGal)', fontsize=20)
plt.legend(loc='best', fontsize=20)
plt.show()
```
