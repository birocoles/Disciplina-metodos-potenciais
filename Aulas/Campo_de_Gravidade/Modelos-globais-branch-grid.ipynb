### Gravidade na sup. da Terra

```python
grav_st = Grid.load('eigen-gl04c-grav-earth.gdf')

print(grav_st.metadata)

#Longitude central
lon_central = 0.5*(grav_st.area[3] + grav_st.area[2])

proj = grav_st.set_projection(crs.Geodetic()).reproject(crs.PlateCarree(central_longitude=lon_central))

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(projection=proj.projection))

states = cf.NaturalEarthFeature(category='cultural', scale='50m', facecolor='none',
                                name='admin_1_states_provinces_shp')
ax.add_feature(states, edgecolor='black')

mpl.title('Gravidade (mGal) na sup. da Terra', fontsize=14)
proj.contourf('gravity_earth', ax=ax)
mpl.gca().coastlines()
gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
mpl.tight_layout(pad=0)
```

### Gravidade normal (mGal) e Anomalia Ar-livre (mGal)

```python
def gravidade_normal(lat):
    lat = (np.pi/180)*lat
    # WGS84 (Hofmann-Wellenhof and H. Moritz, 2006)
    a = 6378137 # metros
    b = 6356752.3142 # metros
    gamma_a = 9.7803253359 # m/s2
    gamma_b = 9.8321849378 # m/s2
    res = (a*gamma_a*np.cos(lat)**2 + b*gamma_b*np.sin(lat)**2)/np.sqrt((a*np.cos(lat))**2 + (b*np.sin(lat))**2)
    # Converter o resultado de m/s2 para mGal
    return utils.si2mgal(res)

def gravidade_normal_em_p(lat, h):
    return gravidade_normal(lat) - 0.3086*h
```

```python
norm=Grid(lon=dist_st.lon, lat=dist_st.lat, shape=dist_st.shape, z=dist_st.h_over_geoid, lonlat=True, projection=crs.Geodetic())

gamma_st = gravidade_normal_em_p(norm.lat, norm.z)
gamma = gravidade_normal(norm.lat)
ar_livre = grav_st.gravity_earth - gamma_st

#Longitude central
lon_central = 0.5*(norm.area[3] + norm.area[2])

norm.add_attribute('gamma', gamma)
norm.add_attribute('gamma_st', gamma_st)
norm.add_attribute('ar_livre', ar_livre)

proj = norm.reproject(crs.PlateCarree(central_longitude=lon_central))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(17, 6), subplot_kw=dict(projection=proj.projection))

ax = axes[0]

states = cf.NaturalEarthFeature(category='cultural', scale='50m', facecolor='none',
                                name='admin_1_states_provinces_shp')
ax.add_feature(states, edgecolor='black')
ax.set_title('Gravidade normal (mGal) na sup. do elipsoide', fontsize=14)
proj.contourf('gamma', ax=ax)
ax.coastlines()
gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
mpl.tight_layout(pad=0)

ax = axes[1]

states = cf.NaturalEarthFeature(category='cultural', scale='50m', facecolor='none',
                                name='admin_1_states_provinces_shp')
ax.add_feature(states, edgecolor='black')
ax.set_title('Gravidade normal (mGal) na sup. da Terra', fontsize=14)
proj.contourf('gamma_st', ax=ax)
#mpl.contourf(proj.lon, proj,lat, proj.gamma_st, vmin=976400, vmax=981000)
ax.coastlines()
gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
mpl.tight_layout(pad=0)
```
