```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
def Legendre_recursivo(t, n):
    if (n == 0):
        return np.ones_like(t)
    if (n == 1):
        return t
    Pn = np.zeros_like(t)
    for i, v in enumerate(t):
        Pn_2 = 1.0
        Pn_1 = v
        for j in range(2,n+1):
            a = (1.0-j)/j
            b = (2.0*j - 1.0)*v/j
            Pn[i] = a*Pn_2 + b*Pn_1
            Pn_2 = Pn_1
            Pn_1 = Pn[i]
    return Pn
```

```python
def Legendre_recursivo2(t, n):
    if (n == 0):
        return 1.0
    if (n == 1):
        return t
    Pn_2 = 1.0
    Pn_1 = t
    for j in range(2,n+1):
        a = (1.0-j)/j
        b = (2.0*j - 1.0)*t/j
        Pn = a*Pn_2 + b*Pn_1
        Pn_2 = Pn_1
        Pn_1 = Pn
    return Pn
```

```python
t = np.linspace(-1.0, 1.0, 50)
t2 = -0.3
PL2 = Legendre_recursivo2(t2, 4)
PL = Legendre_recursivo(t, 3)
plt.close('all')
plt.figure(figsize=(6,6))
plt.axis('scaled')
plt.plot(t, PL, 'ko')
plt.plot([-1,1],[0,0],'k-')
plt.plot([-1,1],[0.5,0.5],'k-')
plt.plot([-1,1],[-0.5,-0.5],'k-')
plt.plot([0,0],[-1,1],'k-')
plt.plot([t2], [PL2], 'ro')
plt.xlim(-1,1)
plt.ylim(-1,1)
#plt.show()
```

```python
r = 1000.0
rlinha = 800.0
psi = 45.0*np.pi/180.0
u = np.cos(psi)

l = np.sqrt(r**2 + rlinha**2 - 2.0*r*rlinha*u)

f = 1.0/l

N = 30
ordem = np.arange(0,N+1,1)
faprox = np.zeros(N+1)
g = 0.0
for n in range(N+1):
    g += (rlinha**n)*Legendre_recursivo2(u, n)/(r**(n+1))
    faprox[n] = g

faprox_max = np.max(faprox)
faprox_min = np.min(faprox)
dfaprox = faprox_max - faprox_min

plt.close('all')
plt.figure(figsize=(6,6))
plt.plot(ordem, faprox, 'r-', linewidth=2)
plt.plot([0,N], [f,f], 'k-', linewidth=2)
plt.xlim(-0.05*N,N+0.05*N)
plt.ylim(faprox_min - 0.05*dfaprox, faprox_max + 0.05*dfaprox)
```
