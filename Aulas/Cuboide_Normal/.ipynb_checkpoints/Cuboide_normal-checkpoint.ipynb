## Modelo Cuboide Normal (MCN)

### Construção do modelo

```python
LC  = 6400000.0 # Metade da extensão da aresta do cubo crosta (C)
LM  = 6330000.0 # Metade da extensão da aresta do cubo manto (M)
LNE = 3400000.0 # Metade da extensão da aresta do cubo nucleo externo (NE)
LNI = 1400000.0 # Metade da extensãoda aresta do cubo nucleo interno (NI)

z0 = 0.0 # Topo do cubo crosta
zc = z0 + LC # Coordenada z do centro de todos os cubos
             # As coordenadas x e y do centro de todos os cubos é zero

rhoC  =  2700.0 # Densidade do crosta
rhoM  =  5000.0 # Densidade do manto
rhoNE = 11000.0 # Densidade do nucleo externo
rhoNI = 13000.0 # Densidade do nucleo interno

drhoM  = rhoM  -  rhoC               # Contraste de densidade entre o manto e a crosta
drhoNE = rhoNE -  rhoM - rhoC        # Contraste de densidade entre o nucleo externo e a crosta
drhoNI = rhoNI - rhoNE - rhoM - rhoC # Contraste de densidade entre o nucleo interno e a crosta

#Massa do MCN
MMCN = (4*LC**3)*(4*LM**3)*(4*LNE**3)*(4*LNI**3)

# Construção do modelo
MCN = [mesher.Prism(-LC, LC, -LC, LC, zc - LC, zc + LC, {'density': rhoC}),          # Cubo crosta
       mesher.Prism(-LM, LM, -LM, LM, zc - LM, zc + LM, {'density': drhoM}),         # Cubo manto
       mesher.Prism(-LNE, LNE, -LNE, LNE, zc - LNE, zc + LNE, {'density': drhoNE}),  # Cubo nucleo interno
       mesher.Prism(-LNI, LNI, -LNI, LNI, zc - LNI, zc + LNI, {'density': drhoNI})]  # Cubo nucleo externo

# Gráfico do perfil vertical de densidade
# O perfil vai da superfície (z = 0) até o centro do modelo (z = zc)
zp  = np.array([zc - LC, zc - LM,
                zc - LM, zc - LNE,
                zc - LNE, zc - LNI,
                zc - LNI, zc])
rho = np.array([rhoC, rhoC,
                rhoM, rhoM,
                rhoNE, rhoNE,
                rhoNI, rhoNI])
mpl.close('all')
mpl.figure(figsize=(8, 8))
mpl.plot(0.001*zp,rho,'k--', linewidth=2)
mpl.xlabel('$z \, (km)$', fontsize=20)
mpl.ylabel('$\\rho \, (kg/m^{3})$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.xlim(0.001*z0, 0.001*zc)
mpl.show()
```

```python
#Visualização 3D do modelo
bounds = [-LC, LC, -LC, LC, z0, (zc + LC)] # Limites da região de visualização 3D
myv.figure()
myv.prisms([MCN[0]], prop='density', opacity = 0.2)
myv.prisms([MCN[1]], prop='density', opacity = 0.4)
myv.prisms([MCN[2]], prop='density', opacity = 0.6)
myv.prisms([MCN[3]], prop='density', opacity = 1.0)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.show()
```

### Plano transversal $yz$ sobre o qual os dados são calculados

```python
ny = 201
nz = 201
shape = (ny,nz)
#areayz = [-LC, zc, -LC, LC] # Limites da área onde os dados serão calculados
areayz = [-100000, 100000, -100000, 100000] # Limites da área onde os dados serão calculados
z, y, x = gridder.regular(areayz, shape, z = 0.0) # Coordenadas z, y e x dos pontos de observação
```

```python
#Visualização 2D do modelo projetado no plano yz
MCN2D = [mesher.Polygon([[-LC, zc - LC], [ LC, zc - LC], [ LC, zc], [-LC, zc]]),
         mesher.Polygon([[-LM, zc - LM], [ LM, zc - LM], [ LM, zc], [-LM, zc]]),
         mesher.Polygon([[-LNE, zc - LNE], [ LNE, zc - LNE], [ LNE, zc], [-LNE, zc]]),
         mesher.Polygon([[-LNI, zc - LNI], [ LNI, zc - LNI], [ LNI, zc], [-LNI, zc]])]

mpl.close('all')
mpl.figure(figsize=(16, 6))
mpl.subplot(1,2,1)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
mpl.ylim(zc, -LC)
mpl.xlim(-LC, LC)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.subplot(1,2,2)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
mpl.ylim(100000, -2000)
mpl.xlim(-51000, 51000)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

```python
#Visualização 3D do modelo
bounds = [-LC, LC, -LC, LC, -LC, (zc + LC)] # Limites da região de visualização 3D
myv.figure()
myv.prisms([MCN[0]], prop='density', opacity = 0.2)
myv.prisms([MCN[1]], prop='density', opacity = 0.4)
myv.prisms([MCN[2]], prop='density', opacity = 0.6)
myv.prisms([MCN[3]], prop='density', opacity = 1.0)
myv.points(np.vstack((x, y, z)).T, color=(1, 0, 0), size=50000, opacity=1, spherical=False)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=4, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.show()
```

### Cálculo do potencial gravitational normal $V_{n}(x,y,z)$ sobre o plano $yz$

```python
Vn = prism.potential(x, y, z, MCN)

mpl.close('all')
mpl.figure(figsize=(16, 6))
mpl.subplot(1,2,1)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
nlevels = mpl.contourf(y, z, Vn, shape, 15)
mpl.colorbar()
mpl.ylim(zc, -LC)
mpl.xlim(-LC, LC)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.subplot(1,2,2)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
mpl.contourf(y, z, Vn, shape, nlevels)
mpl.colorbar()
mpl.ylim(600000, -200000)
mpl.xlim(-500000, 500000)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

### Cálculo do potencial centrífugo $\Phi(x,y,z)$ sobre o plano $yz$

```python
omega = 7292115*(10**-11)
Phi = 0.5*(omega**2)*(y**2)

mpl.close('all')
mpl.figure(figsize=(16, 6))
mpl.subplot(1,2,1)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
nlevels = mpl.contourf(y, z, Phi, shape, 15)
mpl.colorbar()
mpl.ylim(zc, -LC)
mpl.xlim(-LC, LC)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.subplot(1,2,2)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
mpl.contourf(y, z, Phi, shape, nlevels)
mpl.colorbar()
mpl.ylim(900000, -100000)
mpl.xlim(-500000, 500000)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

### Cálculo do potencial de gravidade normal $W_{n}(x,y,z) = V_{n}(x,y,z) + \Phi(x,y,z)$ sobre o plano $yz$

```python
Wn = Vn + Phi

mpl.close('all')
mpl.figure(figsize=(16, 6))
mpl.subplot(1,2,1)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
nlevels = mpl.contourf(y, z, Wn, shape, 15)
mpl.colorbar()
mpl.ylim(zc, -LC)
mpl.xlim(-LC, LC)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.subplot(1,2,2)
for i, p in enumerate(MCN2D):
    mpl.polygon(p, style='-k', linewidth=1)
#mpl.paths([[-LC, z0]], [[LC, z0]], style='r--', linewidth=2)
mpl.contourf(y, z, Wn, shape, nlevels)
mpl.colorbar()
mpl.ylim(600000, -200000)
mpl.xlim(-500000, 500000)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

### Equipotenciais de $W_{n}(x,y,z) = V_{n}(x,y,z) + \Phi(x,y,z)$ sobre o plano $yz$. ## Modelo Cuboide (MC)

### Construção do modelo

```python
# Construção do modelo
MC = list(MCN)
MC2D = list(MCN2D)
```

#### Construção de uma montanha sintética

```python
# Montanha sintética centrada em x = 0 km e y = 0 km
ntopo = 21
Ltopo = 60000.0
nxtopomax =  0.5*Ltopo
nxtopomin = -0.5*Ltopo
nytopomax = -60000.0+Ltopo
nytopomin = -60000.0
ytopo, xtopo = gridder.regular([nytopomin, nytopomax, nxtopomin, nxtopomax],
                               (ntopo,ntopo))
topo = -7000.0*utils.gaussian2d(xtopo, ytopo, 14000.0, 14000.0, x0 = 0.0, y0 = -30000.0)
ptopo = (xtopo == 0.0)
dtopo = Ltopo/(ntopo-1)

# Projeção do modelo no plano yz
for i in range(ntopo):
    ntopoi = nytopomin - 0.5*dtopo + i*dtopo
    MC2D.append(mesher.Polygon([[ntopoi, topo[ptopo][i]], 
                                [ntopoi+dtopo, topo[ptopo][i]], 
                                [ntopoi+dtopo, z0], 
                                [ntopoi, z0]]))

MC_montanha = []
#Modelo 3D
for i, t in enumerate(topo):
    MC_montanha.append(mesher.Prism(xtopo[i]-0.5*dtopo, xtopo[i]+0.5*dtopo, 
                                    ytopo[i]-0.5*dtopo, ytopo[i]+0.5*dtopo, 
                                    t, z0, {'density': rhoC}))
                                    #t, z0, {'density': 0.0}))

MC += MC_montanha

#Massa do MCN
#MMC = MMCN +
```

#### Visualização 2D da montanha

```python
mpl.close('all')
mpl.figure(figsize=(12,12))
mpl.axis('scaled')
for i, p in enumerate(MC2D):
    mpl.polygon(p, style='--k', linewidth=2)
#for i, p in enumerate(MCN2D):
#    mpl.polygon(p, style='-k', linewidth=2)
mpl.contour(y, z, Wn, shape, [Wn0], color='b', linewidth=2, clabel=False)
mpl.ylim(10000, -10000)
mpl.xlim(nytopomin, nytopomax)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

#### Visualização 3D da montanha

```python
bounds = [nxtopomin-0.5*dtopo, nxtopomax+0.5*dtopo, 
          nytopomin-0.5*dtopo, nytopomax+0.5*dtopo, 
          -10000, 10000] # Limites da região de visualização 3D
myv.figure()
for i, m in enumerate(MC_montanha):
    myv.prisms([m], prop='density', opacity = 1.0)
#myv.points(np.vstack((x, y, z)).T, color=(1, 0, 0), size=50000, opacity=1, spherical=False)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=4, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.show()
```

#### Construção de uma bacia sintética

```python
# Bacia sintética centrada em x = 0 km e y = 80 km
nbacia = 21
Lbacia = 60000
nxbaciamax = 0.5*Lbacia
nxbaciamin = -0.5*Lbacia
nybaciamax = Lbacia
nybaciamin = 0.0
ybacia, xbacia = gridder.regular([nybaciamin, nybaciamax, nxbaciamin, nxbaciamax],
                                 (nbacia,nbacia))
bacia = 7000.0*utils.gaussian2d(xbacia, ybacia, 14000.0, 14000.0, x0 = 0.0, y0 = 30000.0)
pbacia = (xbacia == 0.0)
dbacia = Lbacia/(nbacia-1)
```

```python
# Projeção do modelo no plano yz
for i in range(0, nbacia):
    nbaciai = nybaciamin - 0.5*dbacia + i*dbacia
    MC2D.append(mesher.Polygon([[nbaciai, z0], 
                                [nbaciai+dbacia, z0], 
                                [nbaciai+dbacia, bacia[pbacia][i]], 
                                [nbaciai, bacia[pbacia][i]]]))

MC_bacia = []
for i, b in enumerate(bacia):
    MC_bacia.append(mesher.Prism(xbacia[i]-0.5*dbacia, xbacia[i]+0.5*dbacia, 
                                 ybacia[i]-0.5*dbacia, ybacia[i]+0.5*dbacia, 
                                 z0, b, {'density': -100}))

MC += MC_bacia

#Massa do MCN
#MMC = MMCN +
```

#### Visualização 2D da bacia

```python
mpl.close('all')
mpl.figure(figsize=(12,12))
mpl.axis('scaled')
for i, p in enumerate(MC2D):
    mpl.polygon(p, style='--k', linewidth=2)
#for i, p in enumerate(MCN2D):
#    mpl.polygon(p, style='-k', linewidth=2)
mpl.contour(y, z, Wn, shape, [Wn0], color='b', linewidth=2, clabel=False)
mpl.ylim(10000, -10000)
mpl.xlim(nybaciamin, nybaciamax)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```
