# Este notebook ilustra como as equipotenciais de um campo gravitacional se deformam em função de uma redistribuição de massa

```python
#source1: https://neuronvisio.readthedocs.org/en/latest/gettingstarted.html#troubleshooting
#source2: http://ipython.org/ipython-doc/dev/interactive/reference.html#pyqt-and-pyside

import sip
sip.setapi('QString', 2)
sip.setapi('QVariant', 2)
```

```python
%matplotlib inline
import numpy as np
from fatiando.vis import mpl, myv
from fatiando import gridder, mesher, gravmag, utils
from fatiando.gravmag import prism
```

## Cálculo do potencial gravitacional $V_{n}(x,y,z)$ produzido por um cubo homogêneo. Este cubo será denominado *cubo normal*

### Contrução do *cubo normal* com aresta $L$

```python
L = 1000000

#modelo 3D
z0 = -0.5*L
CN = [mesher.Prism(-0.5*L, 0.5*L, -0.5*L, 0.5*L, z0, z0+L, props={'density': 1000})]

MCN = 1000.0*L**3

#modelo 2D
CN2D = [mesher.Polygon([[ 0.5*L, -0.5*L], 
                        [ 0.5*L,  0.5*L], 
                        [-0.5*L,  0.5*L], 
                        [-0.5*L, -0.5*L]])]
```

### Conjunto de pontos sobre os quais o potencial será calculado

```python
lambida = 0.2*L
area = [-lambida+z0, z0+lambida, 
        -lambida, lambida]
shape = (100,100)

z, y, x = gridder.regular(area, shape, z = 0.0)

bounds = [-0.5*L, 0.5*L, -0.5*L, 0.5*L, -0.5*L, 0.5*L] # Limites da região de visualização 3D
myv.figure()
myv.prisms(CN, prop='density', opacity = 0.5)
myv.points(np.vstack((x, y, z)).T, color=(1, 0, 0), size=5000, opacity=1)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.show()
```

### Cálculo do potencial gravitacional $V_{n}(x,y,z)$ produzido pelo *cubo normal*. Por conveniência, $V_{n}(x,y,z)$ será denominado *potencial gravitacional normal*

```python
Vn = prism.potential(x,y,z,CN)
```

```python
mpl.close('all')
mpl.figure(figsize=(8,6))
nlevels=mpl.contourf(y, z, Vn, shape, 30)
mpl.colorbar()
mpl.contour(y, z, Vn, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CN2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()
mpl.show()
```

### As equipontenciais do *cubo normal* são funções da distribuição de massa dentro deste. É possível redistribuir a massa dentro do volume do cubo normal de tal forma que o potencial gravitacional seja igual (ou muito próximo) ao potencial $V_{n}(x,y,z)$.

### Construção de um cubo com distribuição de massa diferente daquela do *cubo normal*, mas que produza um potencial $V_{a}(x,y,z) \approx V_{n}(x,y,z)$

```python
#modelo 3D
CA = list(CN)

l = 0.05*L

X11 = -0.5*l
X12 = 0.5*l
Y11 = -0.5*l
Y12 = 0.5*l
Z11 = z0-l
Z12 = z0

X21 = -0.5*l
X22 = 0.5*l
Y21 = -0.5*l
Y22 = 0.5*l
Z21 = z0+L-l
Z22 = z0+L

CA.append(mesher.Prism(X11, X12, Y11, Y12, Z11, Z12, props={'density': 10000}))
CA.append(mesher.Prism(X21, X22, Y21, Y22, Z21, Z22, props={'density': -10000}))

#modelo 2D
CA2D = list(CN2D)
CA2D.append(mesher.Polygon([[Y11, Z11], 
                            [Y12, Z11], 
                            [Y12, Z12], 
                            [Y11, Z12]]))
CA2D.append(mesher.Polygon([[Y21, Z21], 
                            [Y22, Z21], 
                            [Y22, Z22], 
                            [Y21, Z22]]))

myv.figure()
myv.prisms([CA[0]], prop='density', opacity = 0.3)
myv.prisms([CA[1]], prop='density', opacity = 1)
myv.prisms([CA[2]], prop='density', opacity = 1)
#myv.points(np.vstack((x, y, z)).T, color=(1, 0, 0), size=5000, opacity=1)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.show()
```

### Cálculo do potencial gravitacional $V_{a}(x,y,z)$

```python
Va = prism.potential(x, y, z, CA)
```

```python
mpl.close('all')
mpl.figure(figsize=(15,7), tight_layout=True)
mpl.subplot(1,2,1)
mpl.title('$V_{a} \, (m^{2}/s^{2})$\n', fontsize=20)
nlevels = mpl.contourf(y, z, Va, shape, 30)
mpl.colorbar()
mpl.contour(y, z, Va, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CA2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.subplot(1,2,2)
mpl.title('$V_{n} \, (m^{2}/s^{2})$\n', fontsize=20)
mpl.contourf(y, z, Vn, shape, nlevels)
mpl.colorbar()
mpl.contour(y, z, Vn, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CN2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()
```

```python
mpl.close('all')
mpl.figure(figsize=(8,8))
nlevels = mpl.contour(y, z, Vn, shape, 20, color='r', 
                      linewidth=1, clabel=False, label='$V_{n}$')
mpl.contour(y, z, Va, shape, nlevels, color='k', 
            linewidth=1, clabel=False, label='$V_{a}$')
for i, p in enumerate(CA2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.legend(loc='upper left', fontsize=20)
mpl.m2km()
mpl.show()
```

## Cálculo da atração gravitacional $g$ exercida por todos os modelos

### Conjunto de pontos sobre os quais a aceleração gravitacional será calculada

```python
lambida = 0.2*L
area = [-lambida+z0, z0+lambida, 
        -lambida, lambida]
shape = (100,100)

zp, yp, xp = gridder.regular(area, shape, z = 0.0)
#yp = np.linspace(-lambida, lambida, 100)
#xp = np.zeros_like(yp)
#zp = np.zeros_like(yp) + (z0 - 1.2*l)
```

### Cálculo da aceleração gravitacional

```python
gn = np.sqrt(prism.gx(xp, yp, zp, CN)**2 + \
             prism.gy(xp, yp, zp, CN)**2 + \
             prism.gz(xp, yp, zp, CN)**2)

ga = np.sqrt(prism.gx(xp, yp, zp, CA)**2 + \
             prism.gy(xp, yp, zp, CA)**2 + \
             prism.gz(xp, yp, zp, CA)**2)

gb = np.sqrt(prism.gx(xp, yp, zp, CB)**2 + \
             prism.gy(xp, yp, zp, CB)**2 + \
             prism.gz(xp, yp, zp, CB)**2)

gc = np.sqrt(prism.gx(xp, yp, zp, CC)**2 + \
             prism.gy(xp, yp, zp, CC)**2 + \
             prism.gz(xp, yp, zp, CC)**2)
```

```python
mpl.close('all')
mpl.figure(figsize=(15,7), tight_layout=True)
mpl.subplot(1,2,1)
mpl.title('$g_{a} \, (mGal$)\n', fontsize=20)
nlevels = mpl.contourf(y, z, ga, shape, 30)
mpl.colorbar()
mpl.contour(y, z, ga, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CA2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.subplot(1,2,2)
mpl.title('$g_{n} \, (mGal$)\n', fontsize=20)
mpl.contourf(y, z, gn, shape, nlevels)
mpl.colorbar()
mpl.contour(y, z, gn, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CN2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()
```

```python
mpl.close('all')
mpl.figure(figsize=(15,7), tight_layout=True)
mpl.subplot(1,2,1)
mpl.title('$g_{b} \, (mGal$)\n', fontsize=20)
nlevels = mpl.contourf(y, z, gb, shape, 30)
mpl.colorbar()
mpl.contour(y, z, gb, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CB2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.subplot(1,2,2)
mpl.title('$g_{n} \, (mGal$)\n', fontsize=20)
mpl.contourf(y, z, gn, shape, nlevels)
mpl.colorbar()
mpl.contour(y, z, gn, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CN2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()
```
