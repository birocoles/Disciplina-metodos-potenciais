## Cálculo do potencial gravitacional $V_{n}(x,y,z)$ produzido por um cubo homogêneo. Este cubo será denominado *cubo normal*

### Contrução do *cubo normal* com aresta $L$

```python
L = 1000000

#modelo 3D
z0 = -0.5*L
rhon = 1000
CN = [mesher.Prism(-0.5*L, 0.5*L, -0.5*L, 0.5*L, z0, z0+L, props={'density': rhon})]

MCN = 1000.0*L**3

#modelo 2D
CN2D = [mesher.Polygon([[ 0.5*L, -0.5*L], 
                        [ 0.5*L,  0.5*L], 
                        [-0.5*L,  0.5*L], 
                        [-0.5*L, -0.5*L]])]
```

### Conjunto de pontos sobre os quais o potencial será calculado

```python
lambida = 0.2*L
area = [-lambida+z0, z0+lambida, 
        -lambida, lambida]
shape = (100,100)

z, y, x = gridder.regular(area, shape, z = 0.0)

bounds = [-0.5*L, 0.5*L, -0.5*L, 0.5*L, -0.5*L, 0.5*L] # Limites da região de visualização 3D
myv.figure()
myv.prisms(CN, prop='density', opacity = 0.5)
myv.points(np.vstack((x, y, z)).T, color=(1, 0, 0), size=5000, opacity=1)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.show()
```

### Cálculo do potencial gravitacional $V_{n}(x,y,z)$ produzido pelo *cubo normal*. Por conveniência, $V_{n}(x,y,z)$ será denominado *potencial gravitacional normal*

```python
Vn = prism.potential(x,y,z,CN)
```

```python
mpl.close('all')
mpl.figure(figsize=(8,6))
nlevels=mpl.contourf(y, z, Vn, shape, 30)
mpl.colorbar()
mpl.contour(y, z, Vn, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CN2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()
mpl.show()
```

### As equipontenciais do *cubo normal* são funções da distribuição de massa dentro deste. É possível redistribuir a massa dentro do volume do cubo normal de tal forma que o potencial gravitacional seja igual (ou muito próximo) ao potencial $V_{n}(x,y,z)$.

### Construção de um cubo com distribuição de massa diferente daquela do *cubo normal*, mas que produza um potencial $V_{a}(x,y,z) \approx V_{n}(x,y,z)$

```python
#modelo 3D
CA = list(CN)

la = 0.05*L
rhoa = 10000

X11 = -0.5*la
X12 = 0.5*la
Y11 = -0.5*la
Y12 = 0.5*la
Z11 = z0-la
Z12 = z0

X21 = -0.5*la
X22 = 0.5*la
Y21 = -0.5*la
Y22 = 0.5*la
Z21 = z0+L-la
Z22 = z0+L

CA.append(mesher.Prism(X11, X12, Y11, Y12, Z11, Z12, props={'density': rhoa}))
CA.append(mesher.Prism(X21, X22, Y21, Y22, Z21, Z22, props={'density': -rhoa}))

#modelo 2D
CA2D = list(CN2D)
CA2D.append(mesher.Polygon([[Y11, Z11], 
                            [Y12, Z11], 
                            [Y12, Z12], 
                            [Y11, Z12]]))
CA2D.append(mesher.Polygon([[Y21, Z21], 
                            [Y22, Z21], 
                            [Y22, Z22], 
                            [Y21, Z22]]))

myv.figure()
myv.prisms([CA[0]], prop='density', opacity = 0.3)
myv.prisms([CA[1]], prop='density', opacity = 1)
myv.prisms([CA[2]], prop='density', opacity = 1)
#myv.points(np.vstack((x, y, z)).T, color=(1, 0, 0), size=5000, opacity=1)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.show()
```

### Cálculo do potencial gravitacional $V_{a}(x,y,z)$

```python
Va = prism.potential(x, y, z, CA)
```

```python
mpl.close('all')
mpl.figure(figsize=(15,7), tight_layout=True)
mpl.subplot(1,2,1)
mpl.title('$V_{a} \, (m^{2}/s^{2})$\n', fontsize=20)
nlevels = mpl.contourf(y, z, Va, shape, 30)
mpl.colorbar()
mpl.contour(y, z, Va, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CA2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()

mpl.subplot(1,2,2)
mpl.title('$V_{n} \, (m^{2}/s^{2})$\n', fontsize=20)
mpl.contourf(y, z, Vn, shape, nlevels)
mpl.colorbar()
mpl.contour(y, z, Vn, shape, nlevels, color='k', 
            linewidth=1, clabel=False)
for i, p in enumerate(CN2D):
    mpl.polygon(p, style='-k', linewidth=3)
mpl.xlim(area[2], area[3])
mpl.ylim(area[1], area[0])
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.m2km()
```

### Sejam $\mathbf{a}$ e $\mathbf{b}$ dois vetores. Qual é o resultado de  $\left \| \mathbf{a} + \mathbf{b} \right \| - \left \| \mathbf{a} \right \|$  quando $\left \| \mathbf{a} \right \| \gg \left \| \mathbf{b} \right \|$?

```python
a = np.linspace(0.0, 3000.0, 30)
b = np.array([13.0, 25.0,7.1]).T
bz = '$b_{z}$ = %.3f' % b[2]
nb = '$\mathbf{b}$ = %.3f' % np.linalg.norm(b)

c = np.zeros_like(a)
for i in range (len(a)):
    aux = np.array([0.0,0.0,a[i]]).T + b
    c[i] = np.linalg.norm(aux) - a[i]

mpl.close('all')
mpl.figure(figsize=(8,6))
mpl.title(nb, fontsize=20)
mpl.plot(a, c, 'ro')
mpl.paths([[a[0], b[2]]],[[a[-1], b[2]]], 'k-',
          linewidth=2, label=bz)
mpl.xlim(a[0]-a[-1]*0.05,a[-1]+a[-1]*0.05)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.ylabel('$|| \mathbf{a} + \mathbf{b} || - || \mathbf{a} ||$', 
           fontsize=20)
mpl.xlabel('$|| \mathbf{a} ||$', fontsize=20)
mpl.legend(loc='best', fontsize=20)
mpl.show()
```

### Cálculo da componente $z$ (vertical) da aceleração gravitacional

```python
da = ga - gn
db = gb - gn
dc = gc - gn
```
