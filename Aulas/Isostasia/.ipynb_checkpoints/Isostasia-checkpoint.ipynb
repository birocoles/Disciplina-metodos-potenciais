```python
#source1: https://neuronvisio.readthedocs.org/en/latest/gettingstarted.html#troubleshooting
#source2: http://ipython.org/ipython-doc/dev/interactive/reference.html#pyqt-and-pyside

import sip
sip.setapi('QString', 2)
sip.setapi('QVariant', 2)
```

```python
%matplotlib inline
import numpy as np
from IPython.display import Image
from fatiando.vis import mpl, myv
from fatiando import gridder, mesher, gravmag, utils
from fatiando.gravmag import prism
from fatiando.constants import G, SI2MGAL
```

### Este notebook se utiliza dos conceitos abordados nas aulas [Cuboide](http://nbviewer.ipython.org/github/birocoles/Disciplina-metodos-potenciais/blob/master/Aulas/Cuboide_Normal/Cuboide.ipynb) e [Topografia](http://nbviewer.ipython.org/github/birocoles/Disciplina-metodos-potenciais/blob/master/Aulas/Topografia/Topografia.ipynb)

## Modelo *cubo normal* (CN)

```python
L = 1000000  # aresta do cubo externo
Li = 0.93*L   # aresta do cubo interno

#modelo 3D
z0 = -0.5*L
rho = 1000  # densidade do cubo externo
rhoi = 500  # densidade do cubo interno

CN = [mesher.Prism(-0.5*L,  0.5*L,  -0.5*L,  0.5*L,  -0.5*L,   0.5*L,  props={'density': rho}),
      mesher.Prism(-0.5*Li, 0.5*Li, -0.5*Li, 0.5*Li, -0.5*Li , 0.5*Li, props={'density': rhoi})]

#modelo 2D
CN2D = [mesher.Polygon([[ 0.5*L, -0.5*L], 
                        [ 0.5*L,  0.5*L], 
                        [-0.5*L,  0.5*L], 
                        [-0.5*L, -0.5*L]]),
        mesher.Polygon([[ 0.5*Li, -0.5*Li], 
                        [ 0.5*Li,  0.5*Li], 
                        [-0.5*Li,  0.5*Li], 
                        [-0.5*Li, -0.5*Li]])]

bounds = [-0.5*L, 0.5*L, -0.5*L, 0.5*L, -0.5*L, 0.5*L] # Limites da região de visualização 3D

myv.figure(size=(800,800))
myv.prisms([CN[0]], prop='density', opacity = 0.3)
myv.prisms([CN[1]], prop='density', opacity = 0.7)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.savefig('CN.png')
myv.show()
Image(filename='CN.png')
```

### Coordenadas dos pontos onde os dados serão calculados

```python
lambida = 0.1*L
area = [-lambida, lambida, 
        -lambida, lambida]
shape = (51,51)
dxp = (area[3]-area[2])/(shape[1]-1)
dyp = (area[1]-area[0])/(shape[0]-1)

xp, yp, zp = gridder.regular(area, shape, z = z0-7000)
perfil = (xp == 0.0)
```

### Seção vertical do CN ao longo do perfil x = 0 km, entre as coordenadas y = -100 km e y = 100 km

```python
mpl.close('all')
mpl.figure(figsize=(8,6))
mpl.title('CN ao longo do perfil x = 0 km',
          fontsize=16)
for i, t in enumerate(CN2D):
    mpl.polygon(t)
mpl.paths([[yp[perfil][0], z0]], [[yp[perfil][-1], z0]], 'k--')
mpl.xlim(yp[perfil][0], yp[perfil][-1])
mpl.ylim(z0+0.5*(L-Li)+10000, z0-10000)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

## Modelo *cubo com topografia* e *compensação isostática de Airy* (CTI)

### Construção da topografia

```python
rhot = rho

topo = -6000*utils.gaussian2d(xp, yp, 
                              sigma_x = 30000, sigma_y = 100000, 
                              x0 = 0.0, y0 = 0.0,
                              angle=-45)

zp_sup = topo - 1.0 + z0

CT = []
auxx = 0.5*dxp
auxy = 0.5*dyp
for i, t in enumerate(topo):
    CT.append(mesher.Prism(xp[i]-auxx, xp[i]+dxp-auxx, 
                           yp[i]-auxy, yp[i]+dyp-auxy,
                           z0+t, z0, props={'density': rhot}))
    
CT2D = []

for i, t in enumerate(topo[perfil]):
    CT2D.append(mesher.Polygon([[yp[perfil][i]-auxy, z0], 
                                [yp[perfil][i]+dyp-auxy, z0], 
                                [yp[perfil][i]+dyp-auxy, z0+t], 
                                [yp[perfil][i]-auxy, z0+t]]))
    
mpl.close('all')
mpl.figure(figsize=(16,6))
mpl.subplot(1, 2, 1)
mpl.title('Topografia (km)', fontsize=16)
mpl.contourf(yp, xp, 0.001*topo, shape, 20)
mpl.colorbar()
#cbar.ax.tick_params(labelsize=10) 
mpl.plot(yp, xp, 'ko', markersize=1)
mpl.ylabel('$x \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.subplot(1, 2, 2)
mpl.title('Topografia ao longo do perfil x = 0 km',
          fontsize=16)
for i, t in enumerate(CT2D):
    mpl.polygon(t)
mpl.paths([[yp[perfil][0], z0]], [[yp[perfil][-1], z0]], 'k--')
mpl.xlim(yp[perfil][0], yp[perfil][-1])
mpl.ylim(z0+1000, z0+np.min(topo)-1000)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

```python
bounds = [area[0]-auxx, area[1]+auxx,
          area[2]-auxy, area[3]+auxy,
          z0-7000,z0+1000] # Limites da região de visualização 3D
escala = (1,1,10)

scene = myv.figure(size=(800,600))
myv.prisms(CT, prop='density', opacity = 1, scale=escala)
ax = myv.axes(myv.outline(bounds, scale=escala), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f', )
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
scene.scene.camera.position = [-488377.82089107845, -109890.22530098612, -5193065.2309212582]
scene.scene.camera.focal_point = [2000.0000000015173, 2000.0000000008251, -5029999.9999999655]
scene.scene.camera.view_angle = 30.0
scene.scene.camera.view_up = [0.30138086810157283, 0.065473475009889232, -0.95125327668947979]
scene.scene.camera.clipping_range = [267716.39247072284, 858620.47047816461]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('CT.png')
myv.show()
Image(filename='CT.png')
```

### Construção da compensação isostática (modelo de Airy)

```python
rhoiso = -rhoi

iso = -(rho/rhoi)*topo

CI = []
auxx = 0.5*dxp
auxy = 0.5*dyp
for i, t in enumerate(iso):
    CI.append(mesher.Prism(xp[i]-auxx, xp[i]+dxp-auxx, 
                           yp[i]-auxy, yp[i]+dyp-auxy,
                           z0+0.5*(L-Li), z0+0.5*(L-Li)+t, props={'density': rhoiso}))
    
CI2D = []

for i, t in enumerate(iso[perfil]):
    CI2D.append(mesher.Polygon([[yp[perfil][i]-auxy, z0+0.5*(L-Li)], 
                                [yp[perfil][i]+dyp-auxy, z0+0.5*(L-Li)], 
                                [yp[perfil][i]+dyp-auxy, z0+0.5*(L-Li)+t], 
                                [yp[perfil][i]-auxy, z0+0.5*(L-Li)+t]]))
    
mpl.close('all')
mpl.figure(figsize=(16,6))
mpl.subplot(1, 2, 1)
mpl.title('Modelo de Airy (km)', fontsize=16)
mpl.contourf(yp, xp, 0.001*(z0+0.5*(L-Li)+iso), shape, 20)
mpl.colorbar()
mpl.plot(yp, xp, 'ko', markersize=1)
mpl.ylabel('$x \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.subplot(1, 2, 2)
mpl.title('Modelo de Airy + Topografia em x = 0 km',
          fontsize=16)
for i, t in enumerate(CT2D):
    mpl.polygon(t, style='k-')
for i, t in enumerate(CI2D):
    mpl.polygon(t, style='k-')
mpl.xlim(yp[perfil][0], yp[perfil][-1])
mpl.ylim(z0+0.5*(L-Li)+np.max(iso)+1000, z0+np.min(topo)-1000)
mpl.ylabel('$z \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```

```python
bounds = [area[0]-auxx, area[1]+auxx,
          area[2]-auxy, area[3]+auxy,
          z0+0.5*(L-Li)-1000,z0+0.5*(L-Li)+np.max(iso)+1000] # Limites da região de visualização 3D
escala = (1,1,10)

scene = myv.figure(size=(800,600))
myv.prisms(CI, prop='density', opacity = 1, scale=escala)
ax = myv.axes(myv.outline(bounds, scale=escala), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f', )
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
scene.scene.camera.position = [-446217.4001071119, -415492.84887521144, -4480318.7663311064]
scene.scene.camera.focal_point = [0.0, 0.0, -4590000.0]
scene.scene.camera.view_angle = 30.0
scene.scene.camera.view_up = [-0.11532840617095906, -0.1358951136539715, -0.98398774220771679]
scene.scene.camera.clipping_range = [306294.50779078534, 1015244.5036054436]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('CI.png')
myv.show()
Image(filename='CI.png')
```

```python
#modelo 3D
CTI = list(CN)

for i, t in enumerate(CT):
    CTI.append(t)

for i, t in enumerate(CI):
    CTI.append(t)

#modelo 2D
CTI2D = list(CN2D)

for i, t in enumerate(CT2D):
    CTI2D.append(t)

for i, t in enumerate(CI2D):
    CTI2D.append(t)

bounds = [-0.5*L, 0.5*L, -0.5*L, 0.5*L, -0.5*L, 0.5*L] # Limites da região de visualização 3D

myv.figure(size=(800,800))
myv.prisms([CTI[0]], prop='density', opacity = 0.3)
myv.prisms([CTI[1]], prop='density', opacity = 0.6)
myv.prisms(CT, prop='density', opacity = 1)
myv.prisms(CI, prop='density', opacity = 1)
ax = myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.0f')
ax.axes.x_label, ax.axes.y_label, ax.axes.z_label = 'x (km)', 'y (km)', 'z (km)'
ax.axes.font_factor = 1.0
myv.savefig('CTI.png')
myv.show()
Image(filename='CTI.png')
```

### A figura acima mostra o modelo CT. Este modelo é composto pelo CN (azul transparente) e uma topografia na superfície (prismas azuis sólidos) e um modelo de compensação isostática embaixo (modelo de Airy).

### Cálculo da componente $z$ (vertical) da aceleração gravitacional

```python
gn = prism.gz(xp, yp, zp, CN)
gn_sup = prism.gz(xp, yp, zp_sup, CN)

g0 = prism.gz(xp, yp, zp, CTI)
g0_sup = prism.gz(xp, yp, zp_sup, CTI)
```

### Cálculo do distúrbio gravitacional

```python
d0 = g0 - gn
d0_sup = g0_sup - gn_sup
```

## Efeito da compensação isostática (modelo de Airy)

### Remoção da componente $z$ (vertical) da aceleração gravitacional exercida pelo modelo de compensação isostática CI.

```python
gi = prism.gz(xp, yp, zp, CI)
gi_sup = prism.gz(xp, yp, zp_sup, CI)
```

```python
mpl.close('all')
mpl.figure(figsize=(16,6))
mpl.subplot(1, 2, 1)
mpl.title('Modelo de Airy (km)', fontsize=20)
mpl.contourf(yp, xp, 0.001*(z0+0.5*(L-Li)+iso), shape, 20)
mpl.colorbar()
mpl.plot(yp, xp, 'ko', markersize=1)
mpl.ylabel('$x \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.subplot(1, 2, 2)
mpl.title('$g_{i}$ do modelo de Airy (mGal)', fontsize=20)
mpl.contourf(yp, xp, gi_sup, shape, 20)
mpl.colorbar()
mpl.plot(yp, xp, 'ko', markersize=1)
mpl.ylabel('$x \, (km)$', fontsize=20)
mpl.xlabel('$y \, (km)$', fontsize=20)
mpl.yticks(fontsize=14)
mpl.xticks(fontsize=14)
mpl.m2km()
mpl.show()
```
